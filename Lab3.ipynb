```
import numpy as np
import matplotlib.pyplot as plt

def my_sqrt(n, pre=0.0001):
  neg = n < 0
  if neg:
    raise Exception("please give me a positive number")
  if n == 0:
    return 0
  x_n = n / 2
  while True: # do While
    x_n_1 = 0.5 * (x_n + n / x_n)
    if abs(x_n - x_n_1) < pre:
      return x_n_1
    x_n = x_n_1

x_l = []
my_y_l = []
y_l = []
diff_y = []
for x in np.arange(0, 6, 0.1):
  my_y_l.append(my_sqrt(x))
  y_l.append(np.sqrt(x))
  diff_y.append(my_sqrt(x) - np.sqrt(x))
  x_l.append(x)


plt.plot(x_l, my_y_l)
plt.plot(x_l, y_l)
plt.ylabel('diff of versions')
plt.show()


plt.plot(x_l, diff_y)
plt.ylabel('diff of versions')
plt.show()
# 1e-9 is pretty small :)
```
