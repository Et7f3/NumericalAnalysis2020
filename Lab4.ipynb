```
## Stopping criteria
def gen_cond_arbitrary_number(n_max):
  def cond_arbitrary_number(n, e_n, x_n, f):
    return n <= n_max
  return cond_arbitrary_number

def gen_cond_accuracy_x_n(TOLX):
  def cond_accuracy_x_n(n, e_n, x_n, f):
    return abs(e_n) > TOLX
  return cond_accuracy_x_n

def gen_cond_accuracy_zero(TOLF):
  def cond_accuracy_zero(n, e_n, x_n, f):
    return abs(f(x_n)) > TOLF
  return cond_accuracy_zero
```

### Bisection method algorithm

```
# Defining Function
def f(x):
  return x ** 3 - 5 * x - 9

# Implementing Bisection Method
def bisection(x0, x1, e, condition):
  step = 1
  print('\n\n*** BISECTION METHOD IMPLEMENTATION ***')
  x2 = (x0 + x1) / 2
  while condition(step, (x1 - x0) / 2, x2, f):
    x2 = (x0 + x1) / 2
    print('Iteration-%d, x2 = %0.6f and f(x2) = %0.6f' % (step, x2, f(x2)))
    if f(x0) * f(x2) < 0:
      x1 = x2
    else:
      x0 = x2
    step = step + 1
  print('\nRequired Root is : %0.8f' % x2)

# Input Section
x0 = float(input('First Guess: '))
x1 = float(input('Second Guess: '))
e = float(input('Tolerable Error: '))
max_tries = int(input('Max tries: '))

# Checking Correctness of initial guess values and bisecting
if f(x0) * f(x1) > 0.0:
  print('Given guess values do not bracket the root.')
  print('Try Again with different guess values.')
else:
  bisection(x0, x1, e, gen_cond_arbitrary_number(max_tries))
  bisection(x0, x1, e, gen_cond_accuracy_x_n(e))
  bisection(x0, x1, e, gen_cond_accuracy_zero(e))
```

### Fixed Point Iteration algorithm

```
import math

def f(x):
  return x * x * x + x * x - 1

# Re-writing f(x)=0 to x = g(x)
def g(x):
  return 1 / math.sqrt(1 + x)

# Implementing Fixed Point Iteration Method
def fixedPointIteration(x0, e, N, condition):
  print('\n\n*** FIXED POINT ITERATION ***')
  step = 1
  flag = 1
  x1 = g(x0)
  while condition(step, x1 - x0, x1, f):
    x0 = x1
    x1 = g(x0)
    print('Iteration-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))

    step = step + 1

    if step > N:
      flag = 0
      break

  if flag == 1:
    print('\nRequired root is: %0.8f' % x1)
  else:
    print('\nNot Convergent.')

# Input Section
x0 = float(input('Enter Guess: '))
e = float(input('Tolerable Error: '))
max_tries = int(input('Max tries: '))

fixedPointIteration(x0, e, max_tries, gen_cond_arbitrary_number(max_tries))
fixedPointIteration(x0, e, max_tries, gen_cond_accuracy_x_n(e))
fixedPointIteration(x0, e, max_tries, gen_cond_accuracy_zero(e))
```

### Newtonâ€™s method algorithm

```
# Defining Function
def f(x):
  return x ** 3 - 5 * x - 9
# Defining derivative of function
def g(x):
  return 3 * x ** 2 - 5
# Implementing Newton Method
def newton(x0, e, N, condition):
  print('\n\n*** NEWTON METHOD IMPLEMENTATION ***')
  step = 1
  flag = 1
  x1 = x0 - f(x0) / g(x0)
  while condition(step, x1 - x0, x1, f):
    if g(x0) == 0.0:
      print('Divide by zero error!')
      break

    x0 = x1
    x1 = x0 - f(x0) / g(x0)
    print('Iteration-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))
    step = step + 1

    if step > N:
      flag = 0
      break
  if flag == 1:
    print('\nRequired root is: %0.8f' % x1)
  else:
    print('\nNot Convergent.')

# Input Section
x0 = float(input('Enter Guess: '))
e = float(input('Tolerable Error: '))
max_tries = int(input('Max tries: '))

newton(x0, e, max_tries, gen_cond_arbitrary_number(max_tries))
newton(x0, e, max_tries, gen_cond_accuracy_x_n(e))
newton(x0, e, max_tries, gen_cond_accuracy_zero(e))
```

### Secant method algorithm

```
# Defining Function
def f(x):
  return x ** 3 - 5 * x - 9

# Implementing Secant Method
def secant(x0, x1, e, N, condition):
  print('\n\n*** SECANT METHOD IMPLEMENTATION ***')
  step = 1
  if f(x0) == f(x1):
    print('Divide by zero error!')
    return
  x2 = x0 - (x1 - x0) * f(x0) / (f(x1) - f(x0))
  while condition(step, x1 - x0, x2, f):
    x0 = x1
    x1 = x2
    if f(x0) == f(x1):
      print('Divide by zero error!')
      break
    x2 = x0 - (x1 - x0) * f(x0) / (f(x1) - f(x0))
    print('Iteration-%d, x2 = %0.6f and f(x2) = %0.6f' % (step, x2, f(x2)))
    step = step + 1

    if step > N:
      print('Not Convergent!')
      break
  print('\n Required root is: %0.8f' % x2)

# Input Section
x0 = float(input('First Guess: '))
x1 = float(input('Second Guess: '))
e = float(input('Tolerable Error: '))
max_tries = int(input('Max tries: '))

secant(x0, x1, e, max_tries, gen_cond_arbitrary_number(max_tries))
secant(x0, x1, e, max_tries, gen_cond_accuracy_x_n(e))
secant(x0, x1, e, max_tries, gen_cond_accuracy_zero(e))
```
